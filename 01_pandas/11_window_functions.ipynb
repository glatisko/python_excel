```python
import pandas as pd
from vega_datasets import local_data

seattle = local_data.seattle_weather()
seattle.head()
```

```python
# create LAG variable Lagging one day behind 
# in new columns precip_lag, data shifted one row down
seattle['precip_lag'] = seattle['precipitation'].shift(1)
seattle.head()
```

```python
# shift data two rows up
seattle['precip_lead'] = seattle['precipitation'].shift(-2)
seattle.tail()
```

```python
# if precipitaion varias a lot day by day, to smooth it out, create ROLLING AVERAGE
# create  mean value for 14 days period and assing to row 14
seattle['precip_biweekly'] = seattle['precipitation'].rolling(14).mean()
seattle.head(14)
```

```python
seattle.plot(x='date', y='precipitation',  figsize=(12, 8))
seattle.plot(x='date', y='precip_biweekly', figsize=(12, 8))
```
