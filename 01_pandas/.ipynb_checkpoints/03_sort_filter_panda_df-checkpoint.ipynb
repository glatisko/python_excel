```python
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Add new column
tips['tip_pct'] = tips['tip']/tips['total_bill']
tips.head()
```

```python
# Round data
tips['tip_pct'] = tips['tip_pct'].round(2)
tips.head()
```

```python
# Drop column 
tips.drop('tip_pct', axis='columns', inplace=True)
tips.columns
```

```python
# Sort by multiple columns
# Time asecnding, total bill descending
tips.sort_values(by=['time', 'total_bill'], ascending=[True, False])
```

```python
# Filtering - Check unique dinnners
dinner = tips.query('time == "Dinner"')
dinner['time'].unique()
```

```python
# find biggest transation of the dinnner shift
dinner_sorted = dinner.sort_values(by='total_bill', ascending=False)
dinner_sorted.head(10)
```

```python
# Print only rows where tips > 5
tips.query('tip > 5')
```

```python
# Filter by multiple conditions
# Filtering tip > 5 AND time = 'Dinner'
tips.query('tip > 5 and time == "Dinner"')
```
